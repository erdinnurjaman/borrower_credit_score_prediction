Terlihat pada kolom `CNT_CHILDREN`,`AMT_INCOME_TOTAL`,`AMT_CREDIT` terdapat sebuah outliers atau nilai yang jauh dari keseluruhan data, maka akan membuang baris yang terdapat outliers tersebut.

```python
# Membuat sebuah fungsi untuk membuang baris yang terdapat outliers
def removing_outlier(df,col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    low_limit = q1 - 1.5*iqr
    hig_limit = q3 + 1.5*iqr
    no_outliers = df[(df[col]>low_limit)&(df[col]<hig_limit)]
    return no_outliers
```

# TRAIN TEST SPLIT

```python
# Membagi dataset menjadi Train dan Test data dengan rasio 80:20
X = new_df.drop('TARGET', axis=1)
y = new_df['TARGET']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)
```

```python
X_train.shape
```

```python
# Melakukan scaling pada dataset
from sklearn.preprocessing import StandardScaler,MinMaxScaler

ss = StandardScaler()

X_train_ss = ss.fit_transform(X_train)
X_test_ss = ss.transform(X_test)
```

# MODELING 

## LOGISTIC REGRESSION 

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, auc, classification_report, confusion_matrix, roc_curve
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression(random_state=44)
model_lr.fit(X_train_ss, y_train)
y_pred_lr = model_lr.predict(X_test_ss)
y_pred_train_lr = model_lr.predict(X_train_ss)

print('Accuracy for Test data :', accuracy_score(y_test, y_pred_lr))
print('Accuracy for Train data :', accuracy_score(y_train, y_pred_train_lr))
```

```python
print('F1 score for Test data :', f1_score(y_test, y_pred_lr))
print('F1 score for Train data :', f1_score(y_train, y_pred_train_lr))
```

```python
print(confusion_matrix(y_test, y_pred_lr),'\n')
print(confusion_matrix(y_train, y_pred_train_lr))
```

```python
print(classification_report(y_test, y_pred_lr))
print(classification_report(y_train, y_pred_train_lr))
```

```python
print('ROC AUC Score for Test data :', roc_auc_score(y_test, y_pred_lr))
print('ROC AUC Score for Train data :', roc_auc_score(y_train, y_pred_train_lr))
```

```python
prob_estimates_lr = model_lr.predict_proba(X_test_ss)

fpr_lr, tpr_lr, threshold_lr = roc_curve(y_test, prob_estimates_lr[:, 1])
nilai_auc_lr = auc(fpr_lr, tpr_lr)

plt.plot(fpr_lr, tpr_lr, 'b', label=f'AUC={nilai_auc_lr}')
plt.plot([0,1], [0,1], 'r--', label='Logistic Classifier')

plt.title('ROC: Receiver Operating Characteristic')
plt.xlabel('Fallout or False Positive Rate')
plt.ylabel('Recall or True Positive Rate')
plt.legend()
plt.show()
```
